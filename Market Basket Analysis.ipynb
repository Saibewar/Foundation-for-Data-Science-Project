Creating a group for each product to pivot the dataset based on Item


```python
Raw_data["rank"] = Raw_data.groupby("Item")["PurchaseID"].ngroup()
```

Pivoting the data and assigning NaN as NONE

```python
Raw_data_Pivot = Raw_data.pivot(index='PurchaseID', columns='rank', values='Item')
Raw_data_Pivot

```

```python
Raw_data_Pivot_NONE = Raw_data_Pivot.fillna('NONE')
```

Created a loop to concatenate all the column values in a single column. 
for first column, take only first column,
for last column, take previous columns and current column (dont take comma),
for rest of the columns, previous columns, current column and a comma.


```python
Raw_data_Pivot_NONE.columns[len(Raw_data_Pivot_NONE.columns)-1]
Raw_data_Pivot_NONE['New_column'] =''
for i in Raw_data_Pivot_NONE.columns:
    if i == Raw_data_Pivot_NONE.columns[-1]:
        Raw_data_Pivot_NONE['New_column'] = Raw_data_Pivot_NONE['New_column'] + Raw_data_Pivot_NONE[i]
    elif i==Raw_data_Pivot_NONE.columns[0] :
        Raw_data_Pivot_NONE['New_column'] = Raw_data_Pivot_NONE[i] +","
    else:
        Raw_data_Pivot_NONE['New_column'] = Raw_data_Pivot_NONE['New_column'] + Raw_data_Pivot_NONE[i] +","
Raw_data_Pivot_NONE['New_column']
```

Remove all the None values from last column


```python
Raw_data_Pivot_NONE['New_column_NONE']=Raw_data_Pivot_NONE['New_column'].replace("NONE","",regex=True)
#Raw_data_Pivot_NONE['New_column_NONE']=Raw_data_Pivot_NONE['New_column_NONE'].replace("NONE,","",regex=True)
Raw_data_Pivot_NONE['New_column_NONE']
```

removing blanks and repeated items and creating a list for items.
apyori take data only in thie format.

```python
final_list =[]
for i in Raw_data_Pivot_NONE['New_column_NONE'].values:
       
    a  = i.split(",")
    without_empty_strings = []
    for string in a:
        if (string != ""):
            if (string not  in without_empty_strings):
                without_empty_strings.append(string)
    final_list.append(without_empty_strings)
    
    
final_list
```

using Apriori function, calculated the combination rules by passing support, confidence and max length.

```python
rules = apriori(final_list, min_support = 0.0005, min_confidence = 0.05, min_lift = 3, max_length = 6, target = "rules")
association_results = list(rules)
print(association_results[0])
```

displaying the item combinations


```python
for item in association_results:
    print(item[0])
```

creating a dataset with final output of combinations

```python
Combinations = pd.DataFrame( columns = ['First Item','Suggested Items','Item_groups', 'Support','Confidence','Lift',"ItemCount"])
for item in association_results:
    
    pair = item[0]
    items = [x for x in pair]
    
    #print("Rule : ", items[0], " -> " + str(items[1 : len(items)+1]))
    #print("Support : ", str(item[1]))
    #print("Confidence : ",str(item[2][0][2]))
    #print("Lift : ", str(item[2][0][3]))
    #print("Item Count : " , len(items))
    
    #print("=============================") 
    Combinations =Combinations.append({'First Item' :items[0] ,
                                       'Suggested Items' : (items[1 : len(items)+1]),
                                       "Item_groups" :str(items) ,
                    'Support' : str(item[1]),
                     'Confidence' : str(item[2][0][2]),
                       'Lift' : str(item[2][0][3]),
                       "ItemCount" : len(items)
                  } , 
                    ignore_index=True)


   
```

Converting Support, confidence and lift into Integer

```python
Combinations['Support']=pd.to_numeric(Combinations['Support'], errors='coerce')
Combinations['Confidence']=pd.to_numeric(Combinations["Confidence"], errors='coerce')
Combinations['Lift']=pd.to_numeric(Combinations['Lift'], errors='coerce')
```

Final set of combinations
